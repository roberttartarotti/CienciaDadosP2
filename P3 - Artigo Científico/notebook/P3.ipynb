# Artigo

## MuxViz: a tool for multilayer analysis and visualization of networks
O artigo selecionado foi: **MuxViz: a tool for multilayer analysis and visualization of networks**, que faz uso do software MuxViz para análise de dados procurando analisar e classificar dados, focando em dados de alta variedade e sistemas complexos de temas relacionados à ciência e engenharia. 

## Introdução

No caso de redes biológicas, com conexão de genes, proteínas, neurônios e outros elementos biológicos, é possivel indicar que os componentes são parte de um mesmo conjunto biológico ou que possuem características comuns. 

O uso de redes complexas com múltiplos níveis dos dados oferece alguns desafios com relação à sua visualização, análise e intepretação, especialmente para elementos biológicos, porém este tema está ganhando cada vez mais relevância e estudos, analisando interações negativas e positivas ocorrentes entre os componentes estudados.

## Métodos

O artigo faz referência ao desafio de análise computacional para visualização de dados de múltiplas camadas, que podem ser mostrados através das camadas de forma explícita ou através de coloração das camadas de acordo com os diferentes tipos de relação que representam.

## Análise de redes de múltiplas camadas empíricas.

O artigo faz uso do software MuxViz para analisar e visualizar dados de redes de múltiplas camadas usando diferentes interações para organismos com dados provenientes do dataset [BioGRID](thebiogrid.org), um dataset público com um total de 720 mil interações em diversos segmentos. No caso do artigo, também existe um estudo de centralidade para os elementos, usando o software do MuxViz.

A análise de interesse para este caso é o de Herpes simplex genetic-interation network, cujo resultado obtido no software MuxViz é mostrado na Figura 1.

Na era atual da análise de dados, existe uma grande quantidade de redes de múltiplas camadas, e com o objetivo de evitar perdas de informações importantes ou obter conclusões erradas, o uso de métodos que exploram estruturas de múltiplas camadas se faz importante. No caso do artigo, há a referência de que o software usado, MuxViz, é uma ferramenta de interesse para estes tipos de análise.

# Metodologia e Resultados intermediários

Os procedimentos definidos para realizar o estudo, no Cytoscape, são:

1.   Carregar os dados brutos e definir as colunas de origem (source) e destino (target).
2.   Usando o menu Tools->Analyze Network, considerando grafo não direcionado, gerar uma análise dos dados pelo Cytoscape;
3.   Usando o menu Apps->CyFinder->Community Detection->Edge Betweenness, para um valor mínimo de 2 nós e ordem ascendente, executar o código para criar as comunidades e classificar os dados existentes, que ficariam visíveis através da aba network, à esquerda.
4.   Criando, em Node Table, na aba inferior, uma coluna de formato string, adicionar cores em hexadecimal que cada comunidade irá usar, com o objetivo de poder obter uma visualização das comunidades nos dados. 
5.   Em style, na aba esquerda, usar Fill color, selecionando a coluna de cor criada e o tipo de mapeamento: "Passthrough Mapping".
6.   Análisar o resultado e comparar com o do artigo.
7.   Recriar os passos 1 e 2 e, usando o Apps->CyFinder->Community Detection->Label Propagation, para um valor mínimo de 2 nodes e ordem ascendente, executar o código para criar as comunidades para classificar os dados existentes, que ficariam visíveis através da aba network, à esquerda.
8.   Repetir passos 4, 5 e 6.

Este procedimento foi aplicado para os dados obtidos no banco de dados do artigo para o cenário de Herpes.

Sendo que dois tipos de algoritimo para criação de comunidades serão criados:

* Edge Betweenness: A técnica de edge betweenness verifica o número de menores caminhos que passa por cada nó, baseado no algoritmo Girvan Newman. 

* Label Propagation: Label propagation é um algoritimo de machine learning semi supervisionado que classifica os dados.

Em que os blocos de imagens representam:

A: Comunidade obtida, onde a cor em cada nó representa a comunidade a que ele está assinalado para diferentes situações de análise.

B: Análise considerando centralidade, marcando os 5 nós mais expressivos de acordo com um rank de PageRank Centrality.

C: Visualização da rede colorindo os elementos de acordo com a rede a que eles pertencem. 

D: Grau de complexidade, obtido através de clusterização hierárquica.

E: Medidas de correlação entre layers:
  * À esquerda: mean edge overlap
  * Ao centro: degree–degree Pearson correlation coefficient 
  * À direita: degree–degree Spearman correlation coefficient.

# Resultados Obtidos

Os resultados obtidos são visiveis na Figura 9 para o Edge Betweenness e na Figura 12 para o Label Propagation, ambos apresentam os elementos divididos entre as comunidades. No caso do edge betweenness, a classificação é mais visual, enquanto que para o label propagation, por ser uma técnica semi supervisionada, outros critérios são considerados.

Em uma primeira análise, ambas as técnicas apresentaram resultados igualáveis aos do artigo, no entanto, o resultado do edge betweenness foi mais satisfatório e com uma melhor aproximação, visualmente, de resultados obtidos pelo artigo com o software MuxViz.

# Evolução do Projeto

O projeto se iniciou pela busca de um artigo de interesse, porém após diversas análises foi verificado que o banco de dados disponível iria requerer processamentos adicionais, e por isso houve uma pesquisa e filtragem de artigos com base no dataset disponível.

O artigo selecionado contém diferentes tipos de análises, e portanto um segundo passo foi a filtragem e definição do tema. Ao término, foi definido que seria usado o tema de herpes.

Foi realizado o estudo do artigo em busca de se entender como o processamento de dados foi feito, de modo que fosse possível replicar o resultado.

Uma vez que os dados foram definidos, o Cytoscape foi usado para gerar as comunidades através de 2 técnicas:

1. Edge Betweenness
2. Label Propagation

Ambos os casos foram comparados com os resultados do artigo.


# Discussão

Apesar de o artigo ter usado o software MuxViz e para este projeto ter sido usado o Cytoscape, e as técnicas empregadas terem sido diferentes, os resultados apresentaram uma similaridade considerável.

Das técnicas empregadas, a que apresentou resultado mais próximo do esperado foi edge betweenness, que produziu um resultado visualmente semelhante ao Aggregate do artigo.

# Conclusão

## Destacar as principais conclusões obtidas no desenvolvimento do projeto
*   Com o dataset apropriado, e entendendo como o artigo foi desenvolvido, mesmo usando técnicas e softwares diferentes para definir as comunidades, o resultado obtido se aproxima relativamente bem do artigo original.

## Destacar os principais desafios enfrentados
*   Entender como o procedimento no artigo foi desenvolvido para poder gerar resultados próximos.
*   Encontrar um artigo com um dataset com baixa necessidade de pré-processamento para ser usado pelo cytoscape.
*   Houve a tentativa de levantar a análise para outros datasets usados no artigo, como o de HIV e homo multiplex, porém este, por ter uma quantidade de dados superior, requisitou recursos de sistema não disponiveis, o que ocasionou na não utilização de mais datasets.

## Principais lições aprendidas
*   É de extrema importância entender como o desenvolvimento no artigo foi feito, para que se possa gerar dados próximos aos gerados no artigo, mesmo sendo feito com técnicas e software diferentes.
*   Na busca de um artigo para ser replicado em uma condição em que dados são requeridos, a busca por um artigo com dataset padronizado reduz a quantidade de processamentos extras necessários para obtenção de melhores resultados.
